# 回歸 / Regression


```python
import pandas as pd

# load the training dataset
bike_data = pd.read_csv('data/daily-bike-share.csv')
bike_data.head()
```

```python
bike_data['day'] = pd.DatetimeIndex(bike_data['dteday']).day
bike_data.head(32)
```

```python
numeric_features = ['temp', 'atemp', 'hum', 'windspeed']
bike_data[numeric_features + ['rentals']].describe()
```

統計信息揭示了每一個細微的動態中數據分佈的一些信息，包括觀察數據（有731條記錄）、均值、標準差、平均和狀態以及四分數值（三次值） 25%、50%—— 這數據中的誤差和大約 75%））。

```python
import pandas as pd
import matplotlib.pyplot as plt

# This ensures plots are displayed inline in the Jupyter notebook
%matplotlib inline

# Get the label column
label = bike_data['rentals']


# Create a figure for 2 subplots (2 rows, 1 column)
fig, ax = plt.subplots(2, 1, figsize = (9,12))

# Plot the histogram   
ax[0].hist(label, bins=100)
ax[0].set_ylabel('Frequency')

# Add lines for the mean, median, and mode
ax[0].axvline(label.mean(), color='magenta', linestyle='dashed', linewidth=2)
ax[0].axvline(label.median(), color='cyan', linestyle='dashed', linewidth=2)

# Plot the boxplot   
ax[1].boxplot(label, vert=False)
ax[1].set_xlabel('Rentals')

# Add a title to the Figure
fig.suptitle('Rental Distribution')

# Show the figure
fig.show()
```

圖表顯示，每日出租的數量從 0 到 3,400 多不等。然而，日租的平均（和中的低端）數量更接近，大部分數據在 0 到 2,200 左右範圍內。 值的值在箱線上迅速顯示為小火花，此捕捉是*異常值* - 突然，幾乎所有數據的異常值範圍的異常高或低。

我們可以對數字特徵進行同樣的視覺探索。讓我們為每個創造一個直方圖。

```python
# Plot a histogram for each numeric feature
for col in numeric_features:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    feature = bike_data[col]
    feature.hist(bins=100, ax = ax)
    ax.axvline(feature.mean(), color='magenta', linestyle='dashed', linewidth=2)
    ax.axvline(feature.median(), color='cyan', linestyle='dashed', linewidth=2)
    ax.set_title(col)
plt.show()
```

```python
import numpy as np

# plot a bar plot for each categorical feature count
categorical_features = ['season','mnth','holiday','weekday','workingday','weathersit', 'day']

for col in categorical_features:
    counts = bike_data[col].value_counts().sort_index()
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    counts.plot.bar(ax = ax, color='steelblue')
    ax.set_title(col + ' counts')
    ax.set_xlabel(col) 
    ax.set_ylabel("Frequency")
plt.show()
```

任何分類特徵都顯示出或或少的*均勻*發生（個別類別的行數多人口樣本）。

- **假期**：假期的天數比非假期的天數少得多。
- **工作日**：工作日比非多。
- **weathersit**：大多數日子是類別*1*（晴朗），類別*2*（霧和雲）次之。類別*3*（小雨或雪）天數相對極端，沒有類別*4 *（大雨、冰或霧）天。

現在我們對列中數據的發布有所了解，我們可以開始尋找特徵與我們希望能夠預測的**租金**標籤之間的關係。

對於數字特徵，我們可以創建散點來顯示特徵值和值的交集。

```python
for col in numeric_features:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    feature = bike_data[col]
    label = bike_data['rentals']
    correlation = feature.corr(label)
    plt.scatter(x=feature, y=label)
    plt.xlabel(col)
    plt.ylabel('Bike Rentals')
    ax.set_title('rentals vs ' + col + '- correlation: ' + str(correlation))
plt.show()
```

```python
# plot a boxplot for the label by each categorical feature
for col in categorical_features:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    bike_data.boxplot(column = 'rentals', by = col, ax = ax)
    ax.set_title('Label by ' + col)
    ax.set_ylabel("Bike Rentals")
plt.show()
```

這些圖顯示了某些類別值與租金之間的關係存在一些差異。例如，週末（**工作日** 0 或 6）和工作日（**工作日** 1 至 5）的租金分佈存在明顯差異。同樣，**假期** 和**工作日** 類別也存在顯著差異。有一個明顯的趨勢，與春季和冬季相比，夏季和秋季月份的租金分佈有所不同。 **weathersit** 類別似乎也對租金分配產生了影響。我們為一個月中的某一天創建的 **day** 功能幾乎沒有變化，這表明它可能無法預測租賃數量。

## 訓練



```python
# Separate features and labels
X, y = bike_data[['season','mnth', 'holiday','weekday','workingday','weathersit','temp', 'atemp', 'hum', 'windspeed']].values, bike_data['rentals'].values
print('Features:',X[:10], '\nLabels:', y[:10], sep='\n')
```

分離數據集後，我們現在有一個名為 **X** 的 numpy 數組，其中包含特徵，而 **y** 包含標籤。

我們*可以*使用所有數據訓練模型；但監督學習中的常見做法是將數據分成兩個子集；一個（通常較大）用於訓練模型的集合，以及一個用於驗證訓練模型的較小的“保留”集合。這使我們能夠通過將預測標籤與已知標籤進行比較來評估模型與驗證數據集一起使用時的表現。重要的是*隨機*地分割數據（而不是說，將前 70% 的數據用於訓練並保留其餘數據用於驗證）。這有助於確保兩個數據子集在統計上具有可比性（因此我們使用與訓練數據具有相似統計分佈的數據來驗證模型）。

為了隨機分割數據，我們將使用 **scikit-learn** 庫中的 *​​train_test_split** 函數。這個庫是最廣泛使用的 Python 機器學習包之一

```python
from sklearn.model_selection import train_test_split

# Split data 70%-30% into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)

print ('Training Set: %d rows\nTest Set: %d rows' % (X_train.shape[0], X_test.shape[0]))
```

```python
# Train the model
from sklearn.linear_model import LinearRegression

# Fit a linear regression model on the training set
model = LinearRegression().fit(X_train, y_train)
print (model)
```

### 模型評估

```python
import numpy as np

predictions = model.predict(X_test)
np.set_printoptions(suppress=True)
print('Predicted labels: ', np.round(predictions)[:10])
print('Actual labels   : ' ,y_test[:10])
```

將每個預測與其對應的“基本事實”實際值進行比較並不是確定模型預測效果的非常有效的方法。讓我們看看是否可以通過可視化將預測與實際標籤進行比較的散點圖來獲得更好的指示。我們還將覆蓋一條趨勢線，以大致了解預測標籤與真實標籤的對齊程度。

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.scatter(y_test, predictions)
plt.xlabel('Actual Labels')
plt.ylabel('Predicted Labels')
plt.title('Daily Bike Share Predictions')
# overlay the regression line
z = np.polyfit(y_test, predictions, 1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test), color='magenta')
plt.show()
```

有一定的對角線趨勢，預測值和實際值的交點一般都沿著趨勢線的路徑走；但是由線表示的理想函數與結果之間存在相當大的差異。此方差表示模型的 *殘差 * - 換句話說，當模型將其在訓練期間學到的係數應用於驗證數據時預測的標籤與驗證標籤的實際值之間的差異。當模型與標籤未知的新數據一起使用時，根據驗證數據評估的這些殘差表示預期的*錯誤*水平。

您可以通過計算一些常用的評估指標來量化殘差。我們將重點關注以下三點：

- **均方誤差 (MSE)**：預測值和實際值之間的平方差的平均值。這產生了一個相對度量，其中值越小，模型的擬合越好
- **均方根誤差 (RMSE)**：MSE 的平方根。這會產生與標籤相同單位的絕對指標（在這種情況下，是租賃數量）。值越小，模型越好（簡單來說，它代表了預測錯誤的平均租金！）
- **決定係數（通常稱為 *R-squared* 或 R<sup>2</sup>）**：一個相對度量，其中值越高，模型擬合越好。本質上，該度量表示模型能夠解釋的預測標籤值和實際標籤值之間的差異有多大

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, predictions)
print("MSE:", mse)

rmse = np.sqrt(mse)
print("RMSE:", rmse)

r2 = r2_score(y_test, predictions)
print("R2:", r2)
```
